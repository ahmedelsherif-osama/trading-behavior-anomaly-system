```python
import pandas as pd

historical_df = pd.read_csv("../data/raw/historical_data.csv")
historical_df.shape
```

#### This dataset contains 211,224 trading execution records with 16 attributes describing user-level trading behavior including order execution, size, pricing, and profitability.

```python
historical_df.columns
```

#### Each row represent a single executed trade event.
#### Primary entity appears to be: Account (user-level identifier)

#### Entity Mapping:
    - User ID: Account
    - Asset: Coin
    - Timestamp: Timestamp / Timestamp IST
    - Action: Side, Direction
    - Trade size: Size USD, Size Tokens
    - Outcome: Closed PnL
    - Market context: Execution Price
    - Transaction metadata: Order ID, Trade ID, Transaction Hash

#### Behavioral Features Definition

We define the following user-level behavioral features for anomaly detection:

1. trades_per_day
2. total_trade_volume_per_day
3. avg_trade_size
4. time_between_trades (optional / derived)


#### Feature Feasibility Check

We validate whether the dataset supports computation of the defined behavioral features.

All 4 core behavioral features are directly available from transaction-level fields in the dataset without extra assumptions.

#### Feature <-> Column Mapping

    - trades_per_day -> Account + Timestamp
    - total_trades_volume_per_day -> Size USD + Account + Timestamp
    - avg_trading_size -> Size USD aggregated per Account/Day
    - time_between_trades -> Timestamp deltas per Account
